```python
def r_pay_process(r_id, r_pw, r_pw2):
    login(r_id, r_pw)
    trans1(r_pw2)
    trans2(r_pw2)
    
```

```python
def login(r_id, r_pw):
    driver.get('https://www.t-treeworld.com/login-sam.asp')
    driver.find_element_by_id('userid').send_keys(r_id)
    driver.find_element_by_id('pass').send_keys(r_pw)
    driver.find_element_by_xpath('//input[@class="button"]').click()

    
```

```python
def trans1(r_pw2):
    driver.get('https://www.t-treeworld.com/office/pages/wallet/conversion2.asp')
    value = driver.find_element_by_name('bal_amt_pr').get_attribute('value')
    if int(float(value)) < 500:
        return
    driver.find_element_by_name('AMT').send_keys(value)
    driver.find_element_by_name('pwd1').send_keys(r_pw2)
    
    driver.find_element_by_xpath('//input[contains(@onclick, "field_chk")]').click()    
    alert = driver.switch_to_alert()
    alert.accept()
    time.sleep(2)
    alert = driver.switch_to_alert()
    alert.accept()
```
